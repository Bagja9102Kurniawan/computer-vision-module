## Question 2: Erosion

Read the image ‘wirebond.png’.

```python
img = cv2.imread(path+'wirebond.png')
plt.imshow(img), plt.axis('off')
```

Apply erosion with structuring element shape ellipse and of size 10x10.

## Question 3: Combining dilation and erosion

Read the image ‘shapes.png’.

```python
img = cv2.imread(path+'shapes.png')
plt.imshow(img), plt.axis('off')
```

What can you do to restore the image back to the form shown on right?91362d19-5d4b-4c3e-ab4b-d39819fda23c.jpg

```python
se = cv2.getStructuringElement(cv2.MORPH_ERODE,(10,15))
erosion = cv2.erode(img, se, iterations = 2)
plt.imshow(erosion), plt.axis('off')
```

```python
se = np.ones((25,15),np.uint8)
dilation = cv2.dilate(erosion, se, iterations = 3)
plt.imshow(dilation), plt.axis('off')
```

## Question 4: Opening and Closing

Read the image ‘fingerprint.png’.

```python
img = cv2.imread(path+'fingerprint.png')
plt.imshow(img), plt.axis('off')
```

Clean the fingerprint image to obtain the form shown on the right.

```python
er = np.ones((3,3),np.uint8)
erosion = cv2.morphologyEx(img, cv2.MORPH_OPEN, er)
plt.imshow(erosion), plt.axis('off')
```

```python
dilation = cv2.morphologyEx(erosion, cv2.MORPH_CLOSE, er)
plt.imshow(dilation), plt.axis('off')
```
